```python
import pandas as pd 
import geopandas as gpd
```

```python
station = gpd.read_file('./data/上海充电桩_中促盟_处理后.geojson')
station.rename(columns={'myid':'stationId','经度':'stationLon','纬度':'stationLat'},inplace=True)
station
```

```python
pred_data = pd.read_csv('./output/pred.csv')
pred_data
```

```python
station = station[station['stationId'].isin(pred_data['0.000000000000000000e+00'])]
station = station[['stationId','stationLon','stationLat','充电站类型','交流桩数量','直流桩数量','交流桩总装机功率','直流桩总装机功率']]
station.dropna(subset=['直流桩总装机功率'],inplace=True)
station['直流桩总装机功率'] = station['直流桩总装机功率'].apply(lambda x: float(x.split('-')[0]))
station_dc = station[['stationId','stationLon','stationLat','充电站类型','直流桩数量',
       '直流桩总装机功率']]
station_ac = station[['stationId','stationLon','stationLat','充电站类型','交流桩数量',
       '交流桩总装机功率']]
station_dc = station_dc[station_dc['直流桩数量']>0]
station_ac = station_ac[station_ac['交流桩数量']>0]
```

```python
station_dc['charge_speed_station'] = station_dc['直流桩总装机功率']/station_dc['直流桩数量']
station_ac['charge_speed_station'] = station_ac['交流桩总装机功率']/station_ac['交流桩数量']
station_ac['stationId'] = station_ac['stationId'].apply(lambda x: str(x)+'_ac')
station_dc['stationId'] = station_dc['stationId'].apply(lambda x: str(x)+'_dc')
```

```python
station_ac.rename(columns={'交流桩数量':'capacity'},inplace=True)
station_ac.drop(columns=['交流桩总装机功率'],inplace=True)
station_dc.rename(columns={'直流桩数量':'capacity'},inplace=True)
station_dc.drop(columns=['直流桩总装机功率'],inplace=True)
```

```python
station_all = pd.concat([station_ac,station_dc])
station_all.drop(columns=['充电站类型'],inplace=True)
station_all.to_csv('./output/abm_中促盟充电站数据.csv',index=None)
```
